```python
from IPython.display import clear_output
import random

# Initialize Game:
board = [' ',' ',' ',' ',' ',' ',' ',' ',' ']
game_won = False
turn = 0

def random_first_player():
    global first_player
    global second_player
    global marker
    
    if random.choice(['X', 'O']) == 'X':
        first_player="X"
        second_player="O"
    else:
        first_player="O"
        second_player="X"
    marker = first_player

# Decide who goes first
random_first_player()
marker = first_player

def reset_game():
    global board
    global game_won
    global turn
    
    board = [' ',' ',' ',' ',' ',' ',' ',' ',' ']
    game_won = False
    turn = 0
    random_first_player()


def display_board(board):
    print ' {} | {} | {} '.format(board[0],board[1],board[2])
    print('---|---|---')
    print ' {} | {} | {} '.format(board[3],board[4],board[5])
    print('---|---|---')
    print ' {} | {} | {} '.format(board[6],board[7],board[8])
    

def game_won_check():
    global turn
    global game_won
    
    # Doesn't need to run in the first five turns
    if turn > 4:
        
        three_in_a_row = ['XXX', 'OOO']
        
        # if any one row or column is equal to three of the same marker, 
        # the game is won by that marker.
        
        # ROW 1:
        if reduce((lambda x, y:x+y), board[:3]) in (three_in_a_row):
            game_won = True
        # COL 1:
        if reduce((lambda x, y:x+y), board[::3]) in (three_in_a_row):
            game_won = True
        # ROW 2:
        if reduce((lambda x, y:x+y), board[3:6]) in (three_in_a_row):
            game_won = True
        # COL 2:
        if reduce((lambda x, y:x+y), board[1:8:3]) in (three_in_a_row):
            game_won = True
        # ROW 3:
        if reduce((lambda x, y:x+y), board[6:9]) in (three_in_a_row):
            game_won = True
        # COL 3:
        if reduce((lambda x, y:x+y), board[2:9:3]) in (three_in_a_row):
            game_won = True
        # TOP-LEFT -> BOTTOM-RIGHT
        if reduce((lambda x, y:x+y), board[::4]) in (three_in_a_row):
            game_won = True
        # BOTTOM-LEFT -> TOP-RIGH
        if reduce((lambda x, y:x+y), board[2:8:2]) in (three_in_a_row):
            game_won = True 


def player_choice():
    global marker
    global turn
    
    if turn % 2 == 0:
        marker = first_player   
    else:
        marker = second_player

def place_marker(board,marker,position):
    board[position] = marker

def space_check(position,marker):
    global turn

    if position in [0,1,2,3,4,5,6,7,8]:
        if board[position] in ('X','O'):
            print 'Space is occupied!'
        else:
            place_marker(board,marker,position)
            turn += 1
    else:
        print 'Please enter a valid cell # (1 - 9)'
        display_board(board)
        player_input()
        
        
def play_again():
    play_again_prompt = raw_input('Play Again? Y/N')
    
    if play_again_prompt == 'Y' or play_again_prompt == 'y':
        clear_output()
        reset_game()
        display_board(board)
        player_input()
            
    elif play_again_prompt == 'N'or play_again_prompt == 'n':
        reset_game()
        pass
    else:
        play_again()
    

def player_input():
    global game_won
    global board
        
    while turn < 9:
            if game_won:
                print '{} Won the game!'.format(marker)
                break
            else:
                player_choice()
                position = input("Enter Move for Player {} : ".format(marker)) -1 # Translate space num to human terms.
                clear_output()
                space_check(position,marker)
                game_won_check()
            display_board(board)
    
    if turn == 9:
        game_won_check
        if game_won:
            print '{} Won the game!'.format(marker)
        else:
            print 'Draw!'
            game_won = True
    if game_won:
        play_again()
```

```python
def game_start():
    global board
    
    display_board(board)
    player_input()
        
game_start()
```
