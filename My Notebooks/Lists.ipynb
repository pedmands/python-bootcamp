```python
# Lists support NESTING
l_1 = [1,2,3]
l_2 = [4,5,6]
l_3 = [7,8,9]
```

```python
matrix = [l_1,l_2,l_3]
```

```python
matrix
```

```python
matrix[0]
```

```python
matrix[0][0]
```

```python
# List Comprehensions are AMAZE-BALLS!!
first_col = [row[1] for row in matrix]
# Grabs the number in each list at the given index:
# [2,5,8]
```

```python
# 5 + 2
first_col[1] + first_col[0]
```
